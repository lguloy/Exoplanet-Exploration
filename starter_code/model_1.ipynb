```python
import pandas as pd
import sklearn
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.

#feature descriptions are in https://exoplanetarchive.ipac.caltech.edu/docs/API_kepcandidate_columns.html
selected_features = df[['koi_period','koi_time0bk', 'koi_impact','koi_duration','koi_depth','koi_prad','koi_teq','koi_insol','koi_model_snr'\
                        ,'koi_steff','koi_slogg', 'koi_srad','ra','dec','koi_kepmag']]
```

```python
selected_features
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
 from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(selected_features,df['koi_disposition'])
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

#Convert/Replace
```

```python
X_train_scaled
```

```python
#use to_categorical and encode
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
label_encoder.fit(y_test)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)


from tensorflow.keras.utils import to_categorical
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
#Create a model using SVM
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model.fit(X_train_scaled, encoded_y_train)
```

# Train the Model



```python
print(f"Training Data Score: {model.score(X_train_scaled, encoded_y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, encoded_y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10, 50],
              'gamma': [0.0001, 0.0005, 0.001, 0.005]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, encoded_y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
